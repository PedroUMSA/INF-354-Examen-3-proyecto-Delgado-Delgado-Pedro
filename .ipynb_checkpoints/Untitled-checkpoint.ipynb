```python
#clasificacion de hogos venenoso y comestibles
#
#objetivo
#
#Diseñar un algoritmo de aprendizaje por redes neuronales y arbol de decicion sacando su puntaje score y su matriz de confucion
#
#conjunto de tados de 8215
#tenemos 22 variables independientes y 1 dependiete que indica si es venenoso o es comestible
#
#Este conjunto de datos incluye descripciones de muestras hipotéticas correspondientes a 23 especies de hongos con branquias 
#en el hongo de la familia Agaricus y Lepiota extraídas de la Guía de campo de la Sociedad Audubon sobre hongos norteamericanos 
#(1981). Cada especie se identifica como definitivamente comestible, definitivamente venenosa o de comestibilidad desconocida y 
#no se recomienda. Esta última clase se combinó con la venenosa. La Guía establece claramente que no existe una regla simple 
#para determinar la comestibilidad de un hongo; no hay regla como "folletos tres, déjalo ser" para Poisonous Oak e Ivy.
#
#clase                                       (Poisonous[venenoso]-Edible[comestible])
#forma de tapa                               (bell[campana]-conical[cónica]-convex[convexa]-flat[plana]-knobbed[nudosa]-cap-surface[])
#superficie de la tapa                       (fibrous[fibroso]-grooves[ranuras]-scaly[escamoso]-smooth[liso])
#color de la tapa                            (brown[marrón]-buff[ante]-cinnamon[canela]-gray[gris]-green[verde]-pink[rosa]-purple[morado]-red[rojo]-white[blanco]-yellow[amarillo])
#moretones                                   (bruises[moretones]-no[no])
#olor                                        (almond[almendra]-anise[anís]-creosote[creosota]-fishy[pescado]-foulmusty[asqueroso]-none[ninguno]-pungent[picante]-spicy[picante])
#accesorio branquial                         (attached[adjunto]-descending[descendente]-free[libre]-notched[entallado])
#espaciado branquial                         (close[cerca]-crowded[hacinamiento]-distant[distante])
#tamaño de las branquias                     (broad[ancho]-narrow[estrecho])
#color de las branquias                      (black[negro]-brown[marron]-buff[ante]-chocolate[chocolate]-gray[gris]-green[verde]-orange[naranjado]-pink[rosa]-purple[morado]-red[rojo]-white[blanco]-yellow[amarillo])
#forma de tallo                              (enlarging[agrandamiento]-tapering[afilado])
#raíz del tallo                              (bulbous[bulboso]-club[club]-cup[copa]-equal[igual]-rhizomorphs[rizomorfos]-rooted[enraizado]-missing[faltante])
#superficie del tallo sobre el anillo        (fibrous[fibroso]-scaly[escamoso]-silky[sedoso]-smooth[liso])
#superficie del tallo debajo del anillo      (fibrous[fibroso]-scaly[escamoso]-silky[sedoso]-smooth[liso])
#color del tallo sobre el anillo             (brown[marron],buff[ante],cinnamon[canela],gray[gris],orange[naranja],pink[rosa],red[rejo],white[blanco],yellow[amarillo])
#color del tallo debajo del anillo           (brown[marron],buff[ante],cinnamon[canela],gray[gris],orange[naranja],pink[rosa],red[rojo],white[blanco],yellow[amarillo])
#tipo de velo                                (partial[parcial],universal[universal])
#color del velo                              (brown[marron],orange[naranja],white[blanco],yellow[amarrillo])
#número de timbre                            (none[ninguno],one[uno],two[dos])
#tipo de anillo                              (cobwebby[telaraña],evanescent[evanescente],flaring[abocinado],large[grande],none[ninguno],pendant[colgante],sheathing[revestimiento],zone[zona])
#color de impresión de esporas               (black[negro]-brown[marron]-buff[ante]-chocolate[chocolate]-green[verde]-orange[naranjado]-purple[morado]-white[blanco]-yellow[amarillo]])
#población                                   (abundant[abundante],clustered[agrupada],numerous[numerosa],scattered[dispersa],several[varias],solitary[solitaria])
#habitat                                     (grasses[pastos],leaves[hojas],meadows[prados],paths[caminos],urban[urbano],waste[residuos],woods[bosques])

import numpy as np
import pandas as pd

data = pd.read_csv('datos.csv')
cabeza=["clase","forma de tapa","superficie de la tapa","color de la tapa","moretones","olor","accesorio branquial","espaciado branquial","tamaño de las branquias","color de las branquias","forma de tallo","raíz del tallo","superficie del tallo sobre el anillo","superficie del tallo debajo del anillo","color del tallo sobre el anillo","color del tallo debajo del anillo","tipo de velo","color del velo","número de timbre","tipo de anillo","color de impresión de esporas","población","habitat"]
data.columns=cabeza
```

```python
#En "y" colocamos el tributo dependiente que es la clase comestible o venenoso (variable categorica)
#En "x" colocamos todas las variables independientes que clasifican a "y" (variable categorica)
y=data.iloc[:,0]
x=data.iloc[:,1:]
```

```python
x
```

```python
y
```

```python
#empezamos remplazando los datos nulos de cada columna con el valor mas frecuente de cada columna

from sklearn.impute import SimpleImputer#para datos faltantes
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer=imputer.fit(x)
datos=imputer.transform(x)
x=datos
```

```python
x
```

```python
#Empezamos con el preprocesamiento de de cada columna en "x" codificando cada uno de las caracteristicas
#caracteristica ya que son variables categoricas

#luego en y hacemos lo mismo

from sklearn import preprocessing
for i in range(22):
    lx = preprocessing.LabelEncoder()
    lx.fit(x[:,i])
    x[:,i]=lx.fit_transform(x[:,i])
    
ly=preprocessing.LabelEncoder()
ly.fit(y)
y=ly.fit_transform(y)
```

```python
x
```

```python
y
```

```python
#creamos una columna para casa valor distinto que exista en cada columna marcando con 1 y 0 
#por lo que resulda una matriz dispersa
#tambien lo denominamos variables dummy encoding para codifica una variable categorica en nuevas columnas
from sklearn.preprocessing import OneHotEncoder
onehot=OneHotEncoder(categories='auto', sparse=False)
x=onehot.fit_transform(x)
```

```python
x
```

```python
#estandarizacion de escalas
#para la distancia que existe entre los valores se trata de cambiar las escalas sin afectar la distribucion 
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=scale.fit_transform(x)
```

```python
x
```

```python
#ya tenemos los datos listos para un metodo de analisis de datos

#librerias

#este metodo nos proporciona indices de entrenamiento para dividir lo conjuntos de entrenamiento y prueba
from sklearn.model_selection import StratifiedShuffleSplit
sss = StratifiedShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
sss.get_n_splits(x, y)

#red neuronal para la clasificacion
from sklearn.neural_network import MLPClassifier

#arbol de decicion de clasificaion
from sklearn.tree import DecisionTreeClassifier

#matrix de confucion
from sklearn.metrics import confusion_matrix
```

```python
n=0
for train_index, test_index in sss.split(x, y):
    n+=1
    print('////////////////////',n,'///////////////////')

    X_train, X_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    print('\n Red neuronal')
    
    clf = MLPClassifier(hidden_layer_sizes=(10,10,10), max_iter=300, alpha=0.0001,solver='adam', random_state=21)
    clf.fit(X_train, y_train)
    #pre1=clf.predict_proba(X_test[:1])
    prediccion_y=clf.predict(X_test)
    score=clf.score(X_test, y_test)
    print('\n prediccion y:\n',prediccion_y,'\nscore:\n',score)
    print(confusion_matrix(y_test, prediccion_y))
    print('\nArbol de decisión')
    
    arb = DecisionTreeClassifier(max_depth = 2, random_state = 21)
    arb.fit(X_train, y_train)
    predicion2=arb.predict(X_test)
    score2=arb.score(X_test, y_test)
    
    print('\npredicion y:\n',predicion2,'\nscore :\n',score2)
    print(confusion_matrix(y_test, predicion2))
```
